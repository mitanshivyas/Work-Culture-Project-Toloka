## TASK - 1
Create a Python Script that given a worker id, Calculate earnings per day (The script sent was avg. earnings per day)

```python
# the funtion will take the user id and the date for which we want to query a particular worker's income
```

```python
def query_income(worker_id, date):
  date = pd.to_datetime(date, format='%Y-%m-%d')
  income_data["new_date"] = pd.to_datetime(income_data["date"]).dt.tz_convert(None)
  income = income_data.loc[(income_data['uid'] == worker_id) & (income_data['new_date'] == date), 'income'].sum()
  print('The total earnings of worker - ' + str(worker_id) + ' on ' + str(date) + ' = ' + '\033[1m' + str(income) + '\033[1m' + '$')
```

```python
# finding avg number of scroll counts = total scroll counts/ total number of user
# average scroll counts per user = total scroll counts/ total number of user
def avg_scroll():
  avg_scroll_count = log_data['scroll_count'].sum()/len(log_data['uid'].unique())
  print('Average number of scrolls: '+ '\033[1m' + str(avg_scroll_count))
```

```python
avg_scroll()
```

```python
# finding average number of clicks per country = total number of clicks in each country / total number of users that belong to that country
# for example: average number of clicks for USA = total number of clicks by all users in USA combined / total numbers of users in USA
def clicks_per_country(country):
  new_log = log_data.merge(user_data[['uid', 'country']], how='outer', on='uid')
  log_grouped = new_log.groupby(by="country")['click_count'].sum()
  countries = user_data['country'].value_counts()
  countries.to_dict
  avg_clicks = log_grouped[country] / countries.get(country)
  print("Average number of clicks for " + str(country) + " is:  " + '\033[1m' + str(avg_clicks))
```

```python
clicks_per_country("US")
```

```python
clicks_per_country("EG")
```

```python
# total interactions per country (interations with tasks?) = total number of interactions with task for a country / total number of users in that country
def interactions_per_country(country):
  new_log = log_data.merge(user_data[['uid', 'country']], how='outer', on='uid')
  new_log['url'] = new_log['url'].astype(str)
  new_log['interactions_with_tasks'] = list(map(lambda x: x.startswith('https://toloka.yandex.com/task/'), new_log['url']))
  log_grouped = new_log.groupby(by="country")['interactions_with_tasks'].sum()
  countries = user_data['country'].value_counts()
  countries.to_dict
  interactions = log_grouped[country] / countries.get(country)
  print("Average interactions for " + str(country) + ' is: ' +  '\033[1m' + str(interactions))
```

## TASK - 4
Create a Python Script that given a worker id find the time it took to complete a task (e.g. since it started until it finished)

For this particular task, the issue is that we cannot exactly utilize these timestamps to get the time took by the worker to finish the task. Because for that purpose, we need time when they started the task (which we will get from the timestamp column in the log data) and we also need the end time, i.e., time when the worker submitted the task (which we will get from the task reciepts). However, the timestamp given in log data are, I believe, the number of seconds that has passed since the user started working on toloka or since our tracker started tracking their activity, however, the end time of workers work submission in the task receipt is simply a date. If both of these values are some dates or if both of these values could be acquired from the timestamp then we could simply subtract and provide the time taken by the worker to finish a task. However, to figure that out with timestamp when we have no idea to what date that timestamp relates to is not possible.

So my solution is that we get both the data from the timestamps itself. So the initial timestamp would be when the worker first clicked on a task, and the final timestamp would be the last time worker clicked on that task. I am not sure if that is the most efficient way to do it, but I think that will give us the closest to true values.

```python
def task_duration(link):
  # but first we need to check if the task is even complete or not
  # we can get that information information from task receipts, however, I am not sure as to how we will be combining that data
  start = log_data.where(log_data['url'] == link)['timestamp'].dropna().values.tolist()[0]
  end = log_data.where(log_data['url'] == link)['timestamp'].dropna().values.tolist()[-1]
  duration = end - start
  days = duration*24/86400
  print('The time took for the given task to get completed is: ' +  '\033[1m' + str(days) +  '\033[1m' + ' hours.')
```
